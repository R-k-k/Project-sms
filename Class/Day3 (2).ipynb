## Basic prop

```python
# 1. len - return total elements inside the list
s1 = ["delhi", 2j, 14, -6.4, False]
print( type(s1) )
print(s1 )

print(f"Total Elements = {len( s1 ) }")
print("Total Elements = ", len(s1) )
```

```python
# 2. index - return the element at specified position
# listName[position]

s1 = ["delhi", 2j, 14, -6.4, False]
print(s1 )

# display the 3rd element
print( s1[2] )

# display last element
print( s1[4] )
print( s1[-1] )
print( s1[ len(s1)-1 ] )

```

```python
# 3. slice - return chunk (collection) of elements from the list
# listName[start : end : step]
# end is not included...

'''
Default : 
    step --> 1 (positive)
    start --> 0
    end --> len()

    step --> negative
    start --> -1
    end --> -( len(s1) + 1 )

'''
```

```python
# read the first 3 elements
print( s1[0 : 3 : 1] ) # 0, 0 + 1 = 1, 1 + 1 = 2,  XXXXX 2 + 1 = 3   # 0 1 2
print( s1[0 : 3 ] )
```

```python
# read the entire list using slicing
print( s1[0 : len(s1) : 1]  )
print( s1[0 : len(s1) : ]  )
print( s1[0 :  : ]  )
print( s1[ :  : ]  )
print( s1[ :  : -1]  )
```

```python
s1 = ["delhi", 2j, 14, -6.4, False]
print( s1[ -1 : -(len(s1) + 1) : -1]  ) # -1 -2 -3 -4 -5
```

```python
print( s1[2 : 3] ) # 14
print( s1[1 : 4 : 2] ) # 1, 1 + 2 = 3, XXXXXXXXXXXXX 3 + 2 = 5 
print( s1[-1 : -3 : -1] ) #-1 , -1-1 = -2,  XXXXXXXXXXXXXXXXXx -2 - 1 = -3 

print( s1[1 : 4 : -1] ) # 1 , 1 + (-1) = 0 # 1 2 3
print( s1[-1 : -4 : 1] ) # -1 + 1 = 0


print( s1[-5 : -1 : 1] ) # -5, -5 + 1 = -4, -4 + 1 = -3, -3 + 1 = -2
```

```python
# 6. membership - return TRUE if element is present inside the seq

s1 = ["delhi", 2j, 14, -6.4, False]  # ["delhi", 2j, 14, -6.4, 0]

print( -6.4 in s1 )
print( "False" in s1 )
print( 0 in s1 )
```

```python
# 3. reverse

s1 = ["delhi", 14, 2j, 14, -6.4, False, 14 , 14] 
s1 = s1[ : : -1] # s1.reverse()
s1
```

```python
# 4. sort
# sorts the elements in ascending order by default
# sorts the elements in descending order by using reverse = True


s1 = [ 14, -6.4, False, 14] # 0 = False
print( s1.sort() )
print(s1)
```

```python
s1 = [ 14, -6.4, False, 14] 
s1.sort(reverse=True)
s1
```

```python
# 7. Append
# Insert an element to the END of the list
# default POSITION is always LENGTH
```

```python
s1 = ["delhi", 2j, 14, -6.4, False]
#     0          1  2   3    4

print("Len s1 => ", len(s1))
print(s1)

s1.append(99) # at 5
print("Len s1 => ", len(s1))
print(s1)

s1.append(99) # at 6
print(s1)

s1.append(99)
print(s1)
```

```python
# 8. Insert
# Insert an element at any POSITION
# listName.insert( position, ELEMENT )
```

```python
s1 = ["delhi", 2j, 14, -6.4, False]
#     0          1  2   3    4

print("Len s1 => ", len(s1))
print(s1)

# insert 99 between 1st and 2nd element
# "delhi", 2j
#   0      1
# -->  0 + 1, 1

s1.insert(1, 99)
print(s1)
```

```python
# q. insert 777 after 2j
s1 = ["delhi", 2j, 14, -6.4, False]
#         0    1   2


# s1.insert(2, 777)
# s1

pos = s1.index(2j)
pos

s1.insert( pos+1, 777 )
s1
```
