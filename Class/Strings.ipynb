#### 3. Slicing

- return a chunk of the string between start and end positions
- stringName[start : end : step]
- stringName[start : end ]

- end is not included

- step has default value as 1

- With respect to Positive Step Size ; 
    - end by default is len
    - start by default is 0

- With respect to Negative Step Size ; 
    - end by default is -len
    - start by default is -1


```python
# read only odd positions of the word ???
word[1 : len(word) : 2] # 1, 1 + 2 = 3, 3 + 2 = 5
```

```python
# read first 2 chars
word = "python"
print( word[0 : 2] ) # 0, 0 + 1 = 1, 
```

```python
# return th
word = "python"
print( word[2 : 4] )# 2 3
print( word[-4 : -2] )# 2 3
```

```python
# read wrt step
word = "python"
print( word[0 : 5] ) # [ 0 : 5 : 1]
print( word[0 : 5 : 1] ) 
print( word[0 : 5 : 2] ) # 0, 0 + 2 = 2, 2 + 2 = 4,  XXXX 4 + 2 = 6
```

```python
word = "python"
print( word[0 : len(word) : 1] )
print( word[0 : len(word) : ] )
print( word[0 :  : ] )
print( word[: len(word) : ] )
print( word[ :  : ] )
```
